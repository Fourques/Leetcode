## 回溯法核心步骤：
- 路径：当前已选择的排列（部分结果）。

- 选择列表：还可以选择的数字。

- 结束条件：当路径长度等于数组长度时，得到一个完整排列。

### 全排列

给定一个不含重复数字的数组 nums ，返回其 所有可能的全排列 。你可以 按任意顺序 返回答案。

```python
import itertools

nums = [1,2,3]

def permute_inter(nums):
    return list(itertools.permutations(nums))

permute_inter(nums)
```

```python


def permute(nums):
    res = []
    path = []
    used = [False] * len(nums)

    def backtrack(level=0):  # level 表示递归层数
        print("  " * level + f"等级{level} path={path}, used={used}")
        
        if len(path) == len(nums):
            res.append(path[:])
            print("  " * level + f"找到排列: {path}")
            return

        for i in range(len(nums)):
            if used[i]:
                continue

            # 做选择
            path.append(nums[i])
            used[i] = True
            print("  " * level + f"i={i}, path={path}, used={used}")

            # 进入下一层递归
            backtrack(level+1)

            # 撤销选择
            path.pop()
            used[i] = False
            print("  " * level + f"i={i}, path={path}, used={used} (for 循环继续)")

    backtrack()
    return res

# 测试
permute([1,2,3])
```

### 子集

给你一个整数数组 nums ，数组中的元素 互不相同 。返回该数组所有可能的子集（幂集）。

解集 不能 包含重复的子集。你可以按 任意顺序 返回解集。

```python
nums = [1,2,3]

def subset(nums):
    
    n = len(nums)
    ans = []
    path = []

    def dfs(start):
        ans.append(path[:])
        for i in range(start, n):
            path.append(nums[i])
            dfs(i+1)
            path.pop()

    dfs(0)
    return ans

subset(nums)
```

### 电话号码的字母组合

给定一个仅包含数字 2-9 的字符串，返回所有它能表示的字母组合。答案可以按 任意顺序 返回。

给出数字到字母的映射如下（与电话按键相同）。注意 1 不对应任何字母。

<img src="题库图像/17.png" alt="电话" width="400">

```python
digits = "23"

def phonecall(digits):
    if not digits:
        return []
    mp = {
        '2': "abc", '3': "def", '4': "ghi", '5': "jkl",
        '6': "mno", '7': "pqrs", '8': "tuv", '9': "wxyz"
    }

    ans = []
    path = []

    def dfs(start):
        if start == len(digits):
            ans.append("".join(path))
            return
        for ch in mp[digits[start]]:
            path.append(ch)
            dfs(start + 1)
            path.pop()

    dfs(0)

    return(ans)

phonecall(digits)
```

### 组合总和

给你一个 无重复元素 的整数数组 candidates 和一个目标整数 target ，找出 candidates 中可以使数字和为目标数 target 的 所有 不同组合 ，并以列表形式返回。你可以按 任意顺序 返回这些组合。

candidates 中的 同一个 数字可以 无限制重复被选取 。如果至少一个数字的被选数量不同，则两种组合是不同的。 

对于给定的输入，保证和为 target 的不同组合数少于 150 个。

```python
candidates = [2,3,6,7]
target = 7

def combin(candidates, target):
    n = len(candidates)
    ans = []
    path = []

    def dfs(start):
        if sum(path) == target:
            sort_path = sorted(path)
            if sort_path not in ans:
                ans.append(sort_path[:])
            return
        if sum(path) > target:
            return
        for i in range(0, n):
            path.append(candidates[i])
            dfs(start+1)
            path.pop()

    dfs(0)
    return ans

combin(candidates, target)
```

```python
def combinationSum(candidates, target):
    candidates.sort()
    ans = []
    path = []

    def dfs(start, remain):
        if remain == 0:           # 命中
            ans.append(path[:])
            return
        # 从 start 开始，保证组合非降序，避免重复
        for i in range(start, len(candidates)):
            x = candidates[i]
            if x > remain:        # 剪枝：后面更大，没必要试
                break
            path.append(x)        # 选 x
            dfs(i, remain - x)    # 仍从 i 开始 → 允许重复使用 x
            path.pop()            # 回溯

    dfs(0, target)
    return ans
```

### 括号生成

数字 n 代表生成括号的对数，请你设计一个函数，用于能够生成所有可能的并且 有效的 括号组合。

```python
n = 3

def gen(n):
    res = []
    path = []

    def dfs(open_cnt, close_cnt):
        # 已放满 2n 个字符
        if open_cnt == n and close_cnt == n:
            res.append("".join(path))
            return

        # 放 '('
        if open_cnt < n:
            path.append('(')
            dfs(open_cnt + 1, close_cnt)
            path.pop()

        # 放 ')'
        if close_cnt < open_cnt:
            path.append(')')
            dfs(open_cnt, close_cnt + 1)
            path.pop()

    dfs(0, 0)
    return res
gen(n)
```

### 单词搜索

给定一个 m x n 二维字符网格 board 和一个字符串单词 word 。如果 word 存在于网格中，返回 true ；否则，返回 false 。

单词必须按照字母顺序，通过相邻的单元格内的字母构成，其中“相邻”单元格是那些水平相邻或垂直相邻的单元格。同一个单元格内的字母不允许被重复使用。

```python
from collections import Counter

board = [['A','B','C','E'],['S','F','C','S'],['A','D','E','E']]
word = "ABCCED"

def exist(board, word):
    if not board or not board[0]:
        return False
    m, n = len(board), len(board[0])
    
    cnt_board = Counter(ch for row in board for ch in row)
    cnt_word = Counter(word)
    print(cnt_board)
    print(cnt_word)
    for ch, need in cnt_word.items():
        if cnt_board[ch] < need:
            return False
        
    if cnt_board[word[0]] > cnt_board[word[-1]]:
        word = word[::-1]

    def dfs(r, c, i):
        # i 表示要匹配 word[i]
        if r < 0 or r >= m or c < 0 or c >= n:
            return False
        if board[r][c] != word[i]:
            return False
        if i == len(word) - 1:
            return True

        # 原地标记：将当前字符临时替换为一个不可能出现的占位符
        tmp = board[r][c]
        board[r][c] = '#'

        # 四方向
        ok = (dfs(r+1, c, i+1) or
                dfs(r-1, c, i+1) or
                dfs(r, c+1, i+1) or
                dfs(r, c-1, i+1))

        # 回溯恢复
        board[r][c] = tmp
        return ok

    for r in range(m):
        for c in range(n):
            if board[r][c] == word[0] and dfs(r, c, 0):
                return True
    return False

exist(board, word)

        
```

### 131. 分割回文串

给你一个字符串 s，请你将 s 分割成一些 子串，使每个子串都是 回文串 。返回 s 所有可能的分割方案。

```python
s = "aab"

def part(s):
    n = len(s)
    # 1) 预处理回文表 pal[l][r]：s[l..r] 是否为回文
    pal = [[False] * n for _ in range(n)]
    for r in range(n):
        for l in range(r + 1):
            if s[l] == s[r] and (r - l <= 2 or pal[l + 1][r - 1]):
                pal[l][r] = True
    print(pal)
    # 2) 回溯
    ans, path = [], []

    def dfs(start):
        if start == n:
            ans.append(path[:])
            return
        for end in range(start, n):
            if pal[start][end]:
                path.append(s[start:end + 1])
                dfs(end + 1)
                path.pop()

    dfs(0)
    return ans

part(s)
```

```python
def partition(s):
    def is_pal(i, j):
        # 判断 s[i..j] 是否回文（闭区间）
        while i < j:
            if s[i] != s[j]:
                return False
            i += 1; j -= 1
        return True

    n = len(s)
    ans, path = [], []

    def dfs(start):
        if start == n:
            ans.append(path[:]); return
        for end in range(start, n):
            if is_pal(start, end):
                path.append(s[start:end+1])
                dfs(end + 1)
                path.pop()

    dfs(0)
    return ans
```

### N 皇后

按照国际象棋的规则，皇后可以攻击与之处在同一行或同一列或同一斜线上的棋子。

n 皇后问题 研究的是如何将 n 个皇后放置在 n×n 的棋盘上，并且使皇后彼此之间不能相互攻击。

给你一个整数 n ，返回所有不同的 n 皇后问题 的解决方案。

每一种解法包含一个不同的 n 皇后问题 的棋子放置方案，该方案中 'Q' 和 '.' 分别代表了皇后和空位。

```python
n = 9

def Nqueen(n):
    res = []
    cols = set()        # 已占用的列
    diag1 = set()       # 主对角线 r - c
    diag2 = set()       # 副对角线 r + c
    path = []           # 当前解：每一行放置列号

    def dfs(r: int):
        if r == n:
            # 把列号路径转成字符串棋盘
            board = []
            for c in path:
                row = ['.'] * n
                row[c] = 'Q'
                board.append(''.join(row))
            res.append(board)
            return

        for c in range(n):
            if c in cols or (r - c) in diag1 or (r + c) in diag2:
                continue
            cols.add(c); diag1.add(r - c); diag2.add(r + c)
            path.append(c)

            dfs(r + 1)

            path.pop()
            cols.remove(c); diag1.remove(r - c); diag2.remove(r + c)

    dfs(0)
    return res

Nqueen(n)
```
