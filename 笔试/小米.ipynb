n搜船出发去探索海洋深处。这些船编号从1到n，按升序依次排列，第i艘船的耐久度为a[i]。

海怪按特定顺序对这些船发动了k次攻击，首先，它攻击第一船，然后是最后一艘，接着又改击第一艘，依此类推。

海怪的每次攻击都会使船的耐久度降低1，当船的耐久度降至0时，它就会沉没，不再受到收击(因此该船不再是第一艘或最后一艘，海怪只会攻击尚未沉没的船)。如果所有船都沉没了，海怪就没有可攻击的目标，便会游走。 

例如，如果 n=4，k=5，且a=[1.2.4.3]，将会发生以下情况: 

- 海怪攻击第一艘船，其耐久度变为0，现在 a=[2.4.3] 
- 海怪攻击最后一渡船，现在 a=[2.4.2]; 
- 海怪攻击第一艘船，现在 a=[1.4.2}; 
- 海怪攻击最后一艘船，现在 a=[1.4.1]; 
- 海怪攻击第一艘船，其耐久度受为0，现在 a=[4.1] 请问海怪攻击后有多少艘船沉没了?

```python
import sys

def solve():
    t = int(input().strip())
    out_lines = []

    for _ in range(t):
        n, k = map(int, input().split())
        a = list(map(int, input().split()))

        left_hits = (k + 1) // 2
        right_hits = k // 2

        i, j = 0, n - 1
        sunk = 0

        while i <= j and left_hits > 0:
            if a[i] <= left_hits:
                left_hits -= a[i]
                sunk += 1
                i += 1
            else:
                a[i] -= left_hits
                left_hits = 0

        while j >= i and right_hits > 0:
            if a[j] <= right_hits:
                right_hits -= a[j]
                sunk += 1
                j -= 1
            else:
                a[j] -= right_hits
                right_hits = 0

        out_lines.append(str(sunk))

    print("\n".join(out_lines))

```

```python
import sys

def solve():
    input = sys.stdin.readline
    T = int(input().strip())
    out = []

    for _ in range(T):
        n, k = map(int, input().split())
        h = list(map(int, input().split()))

        # 若能改到只剩 1 座或全部一样高，最小耐久就是 0
        if k >= n - 1:
            out.append("0")
            continue

        lo, hi = 0, 2_000_000_000  # 二分 X
        while lo < hi:
            X = (lo + hi) // 2

            # 计算在耐久 ≤ X 时最多能保留多少个不修改的山峰
            dp = [1] * n
            best = 1
            for i in range(n):
                hi_val = h[i]
                di = 1
                # 尽量接在某个 j<i 后面
                for j in range(i):
                    # |h[i]-h[j]| ≤ X*(i-j) 才能把中间改出一条每步 ≤ X 的路径
                    if abs(hi_val - h[j]) <= X * (i - j):
                        v = dp[j] + 1
                        if v > di:
                            di = v
                dp[i] = di
                if di > best:
                    best = di

            # 需要修改的数量 = n - 能保留的数量
            if n - best <= k:
                hi = X
            else:
                lo = X + 1

        out.append(str(lo))

    print("\n".join(out))

if __name__ == "__main__":
    solve()
```

AIce在家里搭了一个服务器。有一天她发现这台服务器的防火墙防护记录猛增，为了避免服务器变成矿机，Aice使用了一个可以对比两次攻击类型是否一致的分析器对防护记录进行了分析。现在给出分析器判新哪些攻击之间存在差别，请你判断该分析器是否输出了合理的结果。当且仅当对于某三次攻击a,b,c，分析器判断其中某两对攻击类型相同且第三对攻击类型不同，则我们认为分析器没有输出合理的结果.
输入描述：
第一行有一个正整数T(1<=T<=10)，代表测试数据的组数,对于每组测法数据，由三行构成:
第一行有两个正整数n,m(1<=n<=200,1<=m<=n*(n-1)/2)代表攻击次数以及分析器指出的类型存在差别的攻击对的数目接下来两行每行均有m个正整数，第i列的两个数x,y代表分析器认为第x次攻击和第y次攻击的类型存在差别。Aice对每一对政击均运行了分析器，这两行包会分析器认为存在区别的每一对攻击。
输出描述：
对于每组测试数遇，若分析器输出的结果合理，输出Yes，否则输出N0。(注意首字母大写)

```python
import sys
from io import StringIO
input_data = """3
4 4
1 1 4 4
2 3 2 3
3 1
1
2
5 8
1 1 2 5 1 3 2 3
2 4 5 4 3 5 3 4

"""
sys.stdin = StringIO(input_data)


T = int(sys.stdin.readline().strip())
ans = []

for _ in range(T):
    n, m = map(int, sys.stdin.readline().split())

    # 下面两行各有 m 个整数：第 i 列的 (x[i], y[i]) 表示一次“不同类型”的判断
    x = list(map(int, sys.stdin.readline().split()))
    y = list(map(int, sys.stdin.readline().split()))

    # 构建“不同类型”无向图
    diff = [[False] * (n + 1) for _ in range(n + 1)]
    for xi, yi in zip(x, y):
        if 1 <= xi <= n and 1 <= yi <= n and xi != yi:
            diff[xi][yi] = diff[yi][xi] = True

    # 枚举所有三元组，检查“恰好一条不同”的情况
    ok = True
    for i in range(1, n + 1):
        if not ok:
            break
        for j in range(i + 1, n + 1):
            if not ok:
                break
            dij = diff[i][j]
            for k in range(j + 1, n + 1):
                cnt = (1 if dij else 0) + (1 if diff[i][k] else 0) + (1 if diff[j][k] else 0)
                if cnt == 1:   # 存在两同一异 -> 不合理
                    ok = False
                    break

    ans.append("Yes" if ok else "No")

print("\n".join(ans))
```

思路（判定充要条件）

把分析器输出的“不同类型”的二元关系看成无向图中的边（diff），未出现在列表中的对视为“同类型”（same）。
题目要求：只要存在三次攻击 a,b,c，其中两对被判为“同类型”、另一对被判为“不同类型”，就判为不合理。
等价表述：在任意三元组 (i,j,k) 中，“不同类型”的边数 不能正好等于 1。允许的情况只有 0 条（全同）、2 条（两不同一同，对应两者同、第三个不同）、3 条（全都不同）。
因此：对每个三元组统计“不同”边的数量，如果出现恰好为 1 的，输出 No；否则 Yes。

2.这是一个经典的N行M列的二维迷宫，每个格子有一个整数，代表这个格子的“奖励”或“惩罚”。
玩家从最左上角的格子(1，1)出发，目的地是最右下角的格子(N,M)，并且玩家只能向右或向下走。
玩家在游戏开始时积分为0，并且每到一个格子(包括起始位置和终点位置)，都需要把当前积分加上这个格子对应的整数(显然，若整数为正就是“奖励”，若为负就是“惩罚”)。当玩家在任意时刻积分为0或负数时，就输掉了游戏。
马老师是玩迷宫速通的老玩家，他想到:如果格子(1，1)对应的整数是负数，就会在游戏一开始就直接输掉游戏，有辱他的一世英名。幸好，马老师具有高超的编程技巧，一眼就能看出如果他使用黑客技术把玩家初始积分设置为x，就可以通过游戏。
聪明的马老师想考考你，x最小可以是多少。

```python
import sys
from io import StringIO
input_data = """1
2 3
0 1 -3
1 -2 0

"""
sys.stdin = StringIO(input_data)
T = int(sys.stdin.readline().strip())
out = []

for _ in range(T):
    n, m = map(int, sys.stdin.readline().split())
    # 读入 n 行，每行 m 个整数
    a = [list(map(int, sys.stdin.readline().split())) for _ in range(n)]

    # 一维 DP（自底向上），dp[j] 表示到达终点前，进入当前行第 j 列所需的最小积分
    INF = 10**18
    dp = [INF] * (m + 1)
    dp[m - 1] = 1          # 右下角的哨兵，让 (n-1,m-1) 统一用同一转移

    for i in range(n - 1, -1, -1):
        for j in range(m - 1, -1, -1):
            need_next = min(dp[j], dp[j + 1])   # 下 或 右
            need_here = need_next - a[i][j]
            dp[j] = 1 if need_here <= 1 else need_here

    out.append(str(dp[0]))

print("\n".join(out))

```
